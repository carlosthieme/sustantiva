```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='species', data=df, )
plt.show()
```

```python
# Comparación de la longitud del sépalo y el ancho del sépalo
sns.scatterplot(x='sepal_length', y='sepal_width',
                hue='species', data=df)
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
# Comparación de la longitud y el ancho de un pétalo
sns.scatterplot(x='petal_length', y='petal_width',
                hue='species', data=df, )
  
plt.legend(bbox_to_anchor=(1, 1), loc=2)
  
plt.show()
```

```python
# Histogramas
fig, axes = plt.subplots(2, 2, figsize=(10,10))
  
axes[0,0].set_title("Largo Sépalo")
axes[0,0].hist(df['sepal_length'], bins=7)
  
axes[0,1].set_title("Ancho Sépalo")
axes[0,1].hist(df['sepal_width'], bins=5);
  
axes[1,0].set_title("Largo Pétalo")
axes[1,0].hist(df['petal_length'], bins=6);
  
axes[1,1].set_title("Ancho Pétalo")
axes[1,1].hist(df['petal_width'], bins=6);
```

```python
# Gráficos de caja para conocer la distribución
sns.set(style="ticks")
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x='species',y='sepal_length',data=df)
plt.subplot(2,2,2)
sns.boxplot(x='species',y='sepal_width', data=df)
plt.subplot(2,2,3)
sns.boxplot(x='species', y='petal_length', data=df)
plt.subplot(2,2,4)
sns.boxplot(x='species', y='petal_width', data=df)
plt.show()
```

```python
# Trazar el histograma y la función de densidad de probabilidad (PDF)
plot = sns.FacetGrid(df, hue="species", height=3)
plot.map(sns.distplot, "sepal_length").add_legend()
  
plot = sns.FacetGrid(df, hue="species", height=3)
plot.map(sns.distplot, "sepal_width").add_legend()
  
plot = sns.FacetGrid(df, hue="species", height=3)
plot.map(sns.distplot, "petal_length").add_legend()
  
plot = sns.FacetGrid(df, hue="species", height=3)
plot.map(sns.distplot, "petal_width").add_legend()
  
plt.show()
```

```python
# Análisis de bivariantes
sns.set_style("whitegrid")
sns.pairplot(df,hue='species',height=3)
plt.show()
```

```python
# Verificación de Correlación
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.plot()
```

```python
# Manejo de Outliers
import sklearn
from sklearn.datasets import load_boston

# IQR
Q1 = np.percentile(df['sepal_width'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(df['sepal_width'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1

print("Form: ", df.shape)

# Upper bound
upper = np.where(df['sepal_width'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(df['sepal_width'] <= (Q1-1.5*IQR))

# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

print("Nueva Forma: ", df.shape)

sns.boxplot(x='sepal_width', data=df)
```
