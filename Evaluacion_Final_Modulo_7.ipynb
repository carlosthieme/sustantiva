```python
#----------------------------------------------------------------------
# Módulo 8 - Evaluación Final
#
# Carlos Thieme B.
#
#----------------------------------------------------------------------

#------ Ejercicio 1 -----------------
import pandas as pd 
import matplotlib.pyplot as plt 

def procesa_notas(notas):
    fig, ax = plt.subplots()

    notas.plot(kind = 'box', ax = ax)

    plt.xticks([])
     
    plt.title('Notas de los Alumno')
    return ax


notas = [5.0, 7.0, 7.0, 6, 5.5, 5.2, 3.5, 4.7, 3.2, 5.9, 6.8, 5.5, 6.1, 5.8, 1.3, 3.2, 
         5.6, 4.5, 6.6, 7.0, 5.2, 4.0, 4.0, 5.9, 6.1, 6.8, 3.9]

lasnotas = pd.Series(notas)

procesa_notas(lasnotas)
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv('info2.csv')

fig, ax = plt.subplots()

df.Survived.value_counts().plot(kind = "pie", labels = ["Fallecido", "Sobrevivientes"], 
                                title = "Distribución de Fallecidos y Sobrevivientes")
plt.show()
```

```python
df.Age.plot(kind = "hist", title = "Histograma de edades")
plt.show()
```

```python
df.groupby("Pclass").size().plot(kind = "bar", title =  "Num. Personas por Clase")
plt.show()
```

```python
df.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", title = "Fallecidos y Sobrevivientes por Clase")
plt.show()
```

```python
df.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", stacked = True, title = "Fallecidos y Sobrevivientes por Clase")
plt.show()
```
