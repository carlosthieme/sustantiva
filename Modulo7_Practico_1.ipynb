```python
import pandas as pd 
import matplotlib.pyplot as plt 

def evolucion_ventas(ventas, tipo):
    graficos = {'lineas':'line', 'barras':'bar', 'sectores':'pie', 'area':'area'}

    fig, ax = plt.subplots()

    ventas.plot(kind = graficos[tipo], ax = ax)

    plt.title('Evolución Número de ventas')

    return ax

df = pd.Series([11205, 24000, 32255, 52220, 34500, 22300, 15000, 26532, 12044, 
                13400], index = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])

evolucion_ventas(df, 'lineas')
plt.show()

evolucion_ventas(df, 'area')
plt.show()

evolucion_ventas(df, 'barras')
plt.show()

evolucion_ventas(df, 'sectores')
plt.show()
```
